```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("times.csv")

data_failed_1 = data[data['failed_disks'] == 1]
data_failed_2 = data[data['failed_disks'] == 2]

plt.figure(figsize=(5, 3))

plt.plot(data_failed_1['chunk_size'], data_failed_1['time'], marker='o', linestyle='-', label='Failed Disks = 1')
plt.plot(data_failed_2['chunk_size'], data_failed_2['time'], marker='o', linestyle='-', color='r', label='Failed Disks = 2')

plt.xlabel('Chunk Size[byte]')
plt.ylabel('Time[s]')
plt.title('Failed disks')
plt.grid(True)

plt.legend()

# Set x-axis ticks to relevant data points
plt.xticks(sorted(data['chunk_size'].unique()))

plt.tight_layout()
plt.savefig("exp_failed_disk.pdf")
plt.show()
```

```python
data = pd.read_csv("times_file_size.csv")

data_file_sizes = [size for size in data['file_size'].unique()]

plt.figure(figsize=(5, 3))

for size in data_file_sizes:
    data_plot = data[data['file_size'] == size]
    plt.plot(data_plot['chunk_size'], data_plot['time'], marker='o', linestyle='-', label=f'{size}KB')

plt.xlabel('Chunk Size[byte]')
plt.ylabel('Time[s]')
plt.title('File size')
plt.grid(True)

plt.legend()

# Set x-axis ticks to relevant data points
plt.xticks(sorted(data['chunk_size'].unique()))

plt.legend(loc='upper left', bbox_to_anchor=(1, 1.05))


plt.tight_layout()
plt.savefig("exp_file_size.pdf")
plt.show()
```

```python
data = pd.read_csv("times_local_remote.csv")

data_local_0 = data[data['is_local'] == 0]
data_local_1 = data[data['is_local'] == 1]

plt.figure(figsize=(5, 3))

plt.plot(data_local_0['chunk_size'], data_local_0['time'], marker='o', linestyle='-', label='Remote')
plt.plot(data_local_1['chunk_size'], data_local_1['time'], marker='o', linestyle='-', color='r', label='Local')

plt.xlabel('Chunk Size[byte]')
plt.ylabel('Time[s]')
plt.title('Storage location')
plt.grid(True)

plt.legend()

# Set x-axis ticks to relevant data points
plt.xticks(sorted(data['chunk_size'].unique()))

plt.tight_layout()
plt.savefig("exp_local_remote.pdf")
plt.show()
```

```python
data = pd.read_csv("times_add_delete.csv")

data_add = data[data['op'] == "add"]
data_delete = data[data['op'] == "delete"]

plt.figure(figsize=(5, 3))

plt.plot(data_add['chunk_size'], data_add['time'], marker='o', linestyle='-', label='Add')
plt.plot(data_delete['chunk_size'], data_delete['time'], marker='o', linestyle='-', color='r', label='Delete')

plt.xlabel('Chunk Size[byte]')
plt.ylabel('Time[s]')
plt.title('File operations')
plt.grid(True)

plt.legend()

# Set x-axis ticks to relevant data points
plt.xticks(sorted(data['chunk_size'].unique()))

plt.tight_layout()
plt.savefig("exp_add_delete.pdf")
plt.show()
```
